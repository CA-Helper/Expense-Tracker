```python
import pandas as pd 
df = pd.read_csv("data.csv", sep = "\t")
df.head()
```

```python
df.isnull().sum()
```

```python
df['Date'] = pd.to_datetime(df['Date'])
```

```python
df.describe()
```

```python
expense_by_type = df.groupby('Type')['Value'].sum()
expense_by_component = df.groupby('Component')['Value'].sum()
```

```python
import matplotlib.pyplot as plt 
expense_by_type.plot(kind='pie', autopct='%1.1f%%')
#expense_by_component.plot(kind = 'pie', autopct = '%1.1f%%')
```

```python
expense_by_component.plot(kind = 'pie', autopct = '%1.1f%%')
```

```python
# Group expenses by 'year'
expense_by_year = df.groupby('Year')['Value'].sum()

# Create a line chart to visualize expense trends over the years
expense_by_year.plot(kind='line', marker='o')
```
